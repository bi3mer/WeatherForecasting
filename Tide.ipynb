```python
%matplotlib inline
import numpy as np  
import matplotlib.pyplot as plt  
```

```python
def graph(formula, x_range):  
    x = np.array(x_range)  
    y = formula(x)
    
    # graph
    fig = plt.figure()
    ax = fig.gca()
    plt.title("Tide over 24 Hours")
    ax.plot(x, y)
    ax.set_xlabel("Hours")
    ax.set_ylabel("Height")

def my_formula(x):
    # Note: this 10 at the end of the formula can be adjusted 
    #       to change the severity of the tide
    return np.sin((x-6)/3.7)*10

graph(my_formula, range(0, 24))
```
